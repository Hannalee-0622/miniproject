## 강화학습 재구성 프레임워크

* **환경(Environment)**: 문제 도메인의 상태(State)와 전이 규칙
* **에이전트(Agent)**: 의사결정을 내리는 정책
* **행동(Action)**: 연속적·이산적 의사결정
* **보상(Reward)**: 원하는 최종 목표(장기 성과)로 설계

```python
환경 이름: CartPole-v1
역할: 막대가 달린 카트를 시뮬레이션, 물리 엔진(간단한 선형 모델)으로 상태 변화를 계산
관측값(Observation): 4-dimensional 연속 벡터
  1. 카트의 수평 위치
  2. 카트의 속도
  3. 막대의 각도
  4. 막대의 각속도

종료 조건:
  1. 막대 각대가 15를 벗어나거나
  2. 카트 위치가 2.4를 벗어나거나
  3. 혹은 최대 스텝 도달 시
```

```python
에이전트(Agent)
역할: 환경으로부터 관측값을 받아 "어떤 행동(action)"을 취할지 결정

학습 전략:
  - Policy: ε-greedy, Q-learning, DQN, Policy Gradient, ...
  - 함수 근사: 신경망을 이용해 관측 -> 행동 매핑

값 기반(value-based):
- 정의: 상태-행동 쌍(s, a)에 대한 가치함수 Q(s, a)를 학습, 가장 높은 Q 값을 반환하는 행동을 선택
- 특징:
  1. 정책을 직접 학습 X
  2. 주로 이산 행동 공간에서 사용

정책 기반(policy-based):
- 정의: 에이전트가 상태(s)에서 바로 확률적 정책                                       을 파라미터            와 함계 직접 학습
- 특징:
  1. 연속 행동 공간에도 자연스럽게 적용
  2. PPO, TRPO

입력: 4차원 관측값

출력: 0 또는 1 (왼쪽 혹은 오른쪽)
```

```python
행동(Action):
- 공간(Action space): Discrete 2
  1. 0 왼쪽
  2. 1 오른쪽

- 물리 효과: 일정 크기의 힘을 카트에 가해 위치, 속도가 바뀜
```

```python
보상(Reward):
- 기본 보상 함수:
  1. 각 타임스텝마다 +1
  2. 즉, 에피소드가 길어질수록 누적 보상이 커짐

- 목표: 가능한 한 오랫동안 막대가 넘어지지 않도록 에피소드를 유지하여 높은 누적 보상 획
```

## 활용 가능 분야 예시:


### 1. 금융·트레이딩

* **알고리즘 트레이딩**

  * 상태: 주가, 거래량, 기술 지표
  * 행동: 매수·매도·관망
  * 보상: 수익률, Sharpe Ratio 등
* **포트폴리오 최적화**

  * 상태: 각 자산 비중·시장 지표
  * 행동: 자산 비중 재조정
  * 보상: 장기 리스크 조정 수익

---

### 2. 로보틱스·자율주행

* **로봇 팔 조작**

  * 상태: 관절 각도·속도, 비전 입력
  * 행동: 모터 토크 연속 제어
* **자율주행 차량**

  * 상태: 카메라 영상, 라이다, 속도
  * 행동: 가속·제동·조향
* **드론·물류 자동화**

  * 상태: 드론 위치·배터리 잔량
  * 행동: 비행 경로·속도 제어

---

### 3. 추천 시스템·광고 최적화

* **Recommender RL**

  * 상태: 사용자 행동 이력, 컨텍스트
  * 행동: 상품·콘텐츠 추천
  * 보상: 클릭률(CTR), 구매 전환율(CVR)
* **실시간 입찰(Bidding)**

  * 상태: 경매/광고 슬롯 정보
  * 행동: 입찰 가격 결정
  * 보상: 광고 캠페인 ROI

---

### 4. 자원 관리·운영 최적화

* **클라우드 자동 스케일링**

  * 상태: CPU·메모리·트래픽
  * 행동: 인스턴스 수 증감
  * 보상: 운영 비용 vs. 서비스 품질
* **스마트 그리드·에너지 관리**

  * 상태: 수요 예측, 배터리 잔량
  * 행동: 저장·방출 전략
* **제조 공정 최적화**

  * 상태: 생산 속도·품질 지표
  * 행동: 설비 파라미터 조정

---

### 5. 헬스케어·의료

* **치료 플랜 추천**

  * 상태: 환자 바이탈·검사 결과
  * 행동: 약물 용량·스케줄 조정
  * 보상: 치료 효과·부작용 최소화
* **의료 영상 분석 + RL**

  * 상태: MRI·CT 스캔 이미지
  * 행동: 관심 영역(lesion) 탐색 경로

---

### 6. AutoML·하이퍼파라미터 튜닝

* **Neural Architecture Search (NAS)**

  * 상태: 이전 모델 성능 지표
  * 행동: 레이어 구성·채널 크기 선택
* **하이퍼파라미터 최적화**

  * 상태: 현재 하이퍼파라미터 셋팅
  * 행동: 학습률·배치 크기 조정
